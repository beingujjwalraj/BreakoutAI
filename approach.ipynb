```python
import pandas as pd
import requests

def get_option_chain(instrument_name, expiry_date, side):
    url = 'https://api.upstox.com/v2/option/chain'
    params = {
        'instrument_key': instrument_name,
        'expiry_date': expiry_date  # For example, '2025-06-26'
    }
    headers = {
        'Accept': 'application/json',
        'Authorization': f'Bearer {access_token}'  # Ensure access_token is defined
    }

    response = requests.get(url, params=params, headers=headers)
    
    # Create an empty DataFrame with specified columns
    optionchain = pd.DataFrame(columns=[
        'instrument_name',
        'strike_price',
        'side',
        'bid_or_ask'
    ])
    
    # Parse the response data
    option_chain = response.json().get('data', [])
    size = len(option_chain)
    
    for x in range(size):
        instrument_name = option_chain[x].get('underlying_key')
        strike_price = option_chain[x].get('strike_price')
        
        # Determine whether to get highest ask for calls (CE) or highest bid for puts (PE)
        if side == 'CE':
            bid_or_ask = option_chain[x]['call_options']['market_data'].get('ask_price')
        elif side == 'PE':
            bid_or_ask = option_chain[x]['put_options']['market_data'].get('bid_price')
        else:
            print('Invalid side input. Please enter either "PE" or "CE".')
            return optionchain  # Return empty DataFrame if side is invalid

        # Add the row to the DataFrame
        optionchain.loc[len(optionchain.index)] = [
            instrument_name, strike_price, side, bid_or_ask
        ]
    
    return optionchain

# Example usage
```
